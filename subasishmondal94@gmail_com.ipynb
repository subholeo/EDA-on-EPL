**<h4>Question:** What is the average point required to finish within the top 4 in EPL and qualify for the Champions League?</h4>

```python
# List of all the unique 25 seasons: 
Seasons=list(Actual_Data['Season'].unique())

# Assigning a variable for summation:
summ=0

# Applying loop to sum up all season top 4 points: 
for i in Seasons:
    season=Actual_Data[Actual_Data['Season']==i]
    Home_data=season.groupby('HomeTeam')['Home_Team_Point'].sum()
    Away_data=season.groupby("AwayTeam")['Away_Team_Point'].sum()
    table=Home_data+Away_data
    table=table.sort_values(ascending=False)
    summ+=table.iloc[:4].sum()

# Finding average point:    
whole_avg=summ/25
top_four_avg_point=whole_avg/4
round(top_four_avg_point)
```

<h4>Observation:</h4>

- The above findings suggest, to finish in top 4 of EPL and to qualify in Champions League, a team in an average has to gain 77 points or more. 

<a name = Section610></a>

**<h4>Question:** What is the average point required to stay above the bottom 3 and not get relegated from the EPL?</h4>

```python
# List of all the unique 25 seasons: 
Seasons=list(Actual_Data['Season'].unique())

# Assigning a variable for summation:
summ=0

# Applying loop to find average point required to not get relegated: 
for i in Seasons:
    season=Actual_Data[Actual_Data['Season']==i]
    Home_data=season.groupby('HomeTeam')['Home_Team_Point'].sum()
    Away_data=season.groupby("AwayTeam")['Away_Team_Point'].sum()
    table=Home_data+Away_data
    table=table.sort_values(ascending=False)
    summ+=table.iloc[-4:].sum()

# Finding average point:    
whole_avg=summ/25
bottom_four_avg_point=whole_avg/4
round(bottom_four_avg_point)
```

<h4>Observation:</h4>

- The above findings suggest, a team has to gather more than 34 points to avoid relegation in the league.

**<h4>Question:** Which season has the top score?</h4>

```python
# Declaring an empty list: 

top_score_list=[]

# Declaring list of seasons: 
Seasons=list(Actual_Data['Season'].unique())

# Looping to extract the top scores in each season:
for i in Seasons:
    season=Actual_Data[Actual_Data['Season']==i]
    Home_data=season.groupby('HomeTeam')['Home_Team_Point'].sum()
    Away_data=season.groupby("AwayTeam")['Away_Team_Point'].sum()
    table=Home_data+Away_data
    table=table.sort_values(ascending=False)
    top_score_list.append(table.iloc[:1][0])

# List of top score in each season:    
top_score_list

# Converting top score list into pandas series:
top_score_list=pd.Series(top_score_list,name='Top Score')
top_score_list

# Converting season list into pandas series:
Seasons=pd.Series(Seasons,name='Season')
Seasons

# Concatinating both the series to obtain the records in a single dataframe:
top_scores=pd.concat([Seasons,top_score_list],axis=1)
top_scores=top_scores.sort_values('Top Score',ascending=False)
```

```python
round(top_scores['Top Score'].mean())
```

```python
# Setting season as the index in the dataframe:
top_scores.set_index('Season',inplace=True)

# Plotting the bar chart to visualize the data:
top_scores.plot.bar(figsize=(10,5))

#Adding some cosmetics to the chart:
plt.xlabel('Season',size=14)
plt.ylabel('Top Score',size=14)
plt.title('Top score per season',size=14)
plt.xticks(rotation='60')
```

```python
# Extracting the points garnered by a particular team in all the 25 seasons:

#List of seasons
Seasons=list(Actual_Data['Season'].unique())

# Declaring a list:
league_points=[]

# Looping to store the points gathered in each season by a particular team:
for i in Seasons:
    season=Actual_Data[Actual_Data['Season']==i]
    Home_data=season.groupby('HomeTeam')['Home_Team_Point'].sum()
    Away_data=season.groupby("AwayTeam")['Away_Team_Point'].sum()
    table=Home_data+Away_data
    point=table.loc['Liverpool']
    league_points.append(point)

# Converting the list in pandas series:
league_points=pd.Series(league_points,name='League Point')
league_points

Seasons=pd.Series(Seasons,name='Season')

# Forming a dataframe and plotting the chart:
pd.concat([Seasons,league_points],axis=1).set_index('Season').plot.bar(figsize=(15,5))

#Adding some cosmetics to the chart:
plt.xlabel('Season',size=14)
plt.ylabel('Total Point',size=14)
plt.title('Season wise total points gathered by Liverpool FC',size=14)
```
